The company's objective is to form a downstream pipeline to determine whether a house is worth investing in or not. Our machine learning component will be one part of a larger data pipeline.

A pipeline (or data pipeline) is basically a sequence of components, where one component's output is fed into the next component as input.

#### Knowing this information you have gathered, what would your problem be? Is it a regression problem or a classification probem? Should we use batch learning or online learning?

1. First of all, we have the labels of the houses. Therefore, we can conclude that it's a _**supervised learning problem**_.

2. Next, we can deduce that it's a _**regression**_ problem and not a classification problem because we're trying to estimate the (continuous) value of samples rather than (discrete) classes.

3. Last, we should be fine using a _**batch learning**_ system and not an online one because we're not going to be consistently receiving a flow of data and therefore don't need to frequently update our model.

## 2-2. Select a Performance Measure

_Next, we have to select a performance measure. This is going to be used to determine how well our model is performing._

Typically, the _**Root Mean Square Error (RMSE)**_ is used as the performance measure. The mathematical equation is as follows:

$$\text{RMSE}(\mathbf{X},\ h) = \sqrt{\frac{1}{m}\sum_{i = 1}^m(h(\vec{x}^{(i)}) - y^{(i)})^2}$$

* $\vec{x}$ or $\mathbf{x}$ $\rightarrow$ one single sample (in this case house)
  * $\vec{x}^{(3)}$ $\rightarrow$ this means the 3rd sample
* $\mathbf{X} \rightarrow$ matrix containing all samples $\vec{x}$
* $m \rightarrow$ number of samples
* $h \rightarrow$ our system's prediction function (a.k.a. hypothesis)
* $y^{(i)} \rightarrow$ the label (price in this case) for the $i$th sample

Basically, we're taking the square of the difference between our prediction and the actual price, then dividing it by the total number of samples.

The RMSE is also referred to as the _Euclidean norm_ (i.e. distance formula). It's also called the $\ell_2$ norm ($\Vert\ \Vert_2$ or $\Vert\ \Vert$).

<br>

Another well-known performance measure is the _**Mean Absolute Error (MAE)**_. The math equation is as follows:

$$\text{MAE}(\mathbf{X},\ h) = \frac{1}{m}\sum_{i = 1}^m\bigg|\ h(\vec{x}^{(i)}) - y^{(i)})\ \bigg|$$

The symbols used in the notation are the same as the RMSE. The notable difference here is that we don't square and take the square root, but we simply take the absolute value of the difference between our prediction and the actual value.

The MAE is also called the $\ell_1$ norm ($\Vert\ \Vert_1$).

<br>

In general, the $\ell_k$ norm of a vector $\vec{x}$ containing $n$ elements is defined as:

$$\Vert \vec{x} \Vert_k = (\vert x_1 \vert^k + \vert x_2 \vert^k + \ldots + \vert x_n \vert^k)^{1/k}$$

### The higher the norm index, the more it focuses on large values and neglects small ones. This is why RMSE is more sensitive to outliers than MAE (because it focuses on large values, it focuses more on points that are further away from the actual label than others).

The fact that the RMSE focuses more on outliers isn't necessarily a bad thing, and the RMSE is actually generally preferred.

````python
import os
import tarfile
from six.moves import urllib


DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if (not os.path.isdir(housing_path)):
        print("Directory does not exist. Creating directory...\n")
        os.makedirs(housing_path)
    else:
        print("Directory already exists.\n")
        pass
    
    tgz_path = os.path.join(housing_path, "housing.tgz")
    print("tgz_path = {}\n".format(tgz_path))
    urllib.request.urlretrieve(housing_url, tgz_path)
    print("urllib.request.urlretrieve(housing_url, tgz_path) = {}\n".format(urllib.request.urlretrieve(housing_url, tgz_path)))
    housing_tgz = tarfile.open(tgz_path)
    print("housing_tgz = {}\n".format(housing_tgz))
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
    
    
fetch_housing_data()
````

## 3-4. Create a Test Set

#### In machine learning, we typically have different sets of data within the same dataset. We usually have a training set and a test set (and a cross-validation set). We use the training data to train our model, and the test data to evaluate how well our model is performing.

##### We want to do this to prevent our system from overfitting.

##### Usually library functions are used, but here the author wrote his own functions.

<br><br>

### I'll use both the written function and the library function.

````python
# Using written function.

import numpy as np


def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]


train_set, test_set = split_train_test(housing_data, 0.25)
print("len(train_set) = {}".format(len(train_set)))
print("len(test_set) = {}".format(len(test_set)))
print("len(test_set) / len(train_set) = {0:.2f}".format(len(test_set) / len(train_set)))
print("\nlen(train_set) + len(test_set) = {}".format(len(train_set) + len(test_set)))
print("len(data) = {}".format(len(housing_data)))
````

````python
# Using library.

from sklearn.model_selection import train_test_split


train_set2, test_set2 = train_test_split(housing_data, test_size=0.25, random_state=42)

print("len(train_set2) = {}".format(len(train_set2)))
print("len(test_set2) = {}".format(len(test_set2)))
print("len(test_set2) / len(train_set2) = {0:.2f}".format(len(test_set2) / len(train_set2)))
print("\nlen(train_set2) + len(test_set2) = {}".format(len(train_set2) + len(test_set2)))
print("len(data) = {}".format(len(housing_data)))
````

<br>
### We notice that both methods produce same results.

#### Until now, we used purely random sampling. This is usually okay, but if the sample size is small then we may introduce sampling bias.

##### For example, when a survey company calls 1,000 people to survey, they don't just randomly select 1,000 people. They try to select the people to represent the entire population.

### We may want to use a technique called [_stratified sampling_](https://en.wikipedia.org/wiki/Stratified_sampling). The pros and cons are nicely explained [here](https://www.investopedia.com/ask/answers/041615/what-are-advantages-and-disadvantages-stratified-random-sampling.asp).

#### We learn that the median income is a very important attribute in predicting the median housing prices, and so decide to create several _strata_ based on that attribute. After taking a look at the histogram for median income above, we notice that most samples' median incomes are clustered from \$20,000 to \$50,000.

##### The following code creates a new attribute called `income_category`by dividing the median income by 1.5 and rounding up using [`numpy.ceil()`](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.ceil.html).

#### Reference:

[`numpy.ceil()`](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.ceil.html)

[`pandas.DataFrame.where()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.where.html)

<br>

````python
import numpy as np


housing_data['income_cat'] = np.ceil(housing_data['median_income'] / 1.5)
housing_data['income_cat'].where(cond=(housing_data['income_cat'] < 5), other=5.0, inplace=True)
````

````python
housing_data['income_cat'].hist()
plt.show()
````

<br>
#### After using this technique and creating the different subcategories (i.e. strata) for the income, we can now stratified sampling rather than normal random sampling.

##### Reference:

[`sklearn.model_selection.StratifiedShuffleSplit`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedShuffleSplit.html)
<br>

````python
from sklearn.model_selection import StratifiedShuffleSplit


split = StratifiedShuffleSplit(n_splits=1, test_size=0.25, random_state=42)

for train_index, test_index in split.split(housing_data, housing_data['income_cat']):
    print("train_index = {}".format(train_index))
    print("test_index = {}".format(test_index))
    strat_train_set = housing_data.loc[train_index]
    strat_test_set = housing_data.loc[test_index]
    print()
    
print("len(strat_train_set) = {}".format(len(strat_train_set)))
print("len(strat_test_set) = {}".format(len(strat_test_set)))
print("len(strat_test_set) / len(strat_train_set) = {0:.2f}".format(len(strat_test_set) / len(strat_train_set)))
````

### Now that we've performed stratified sampling, let's see if our sampling really is representative of our population.

#### In order to do this, take the proportion of how many of each income category was sampled from the entire population.

##### Reference

[`pandas.DataFrame.value_counts()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html)

[`pandas.DataFrame.sort_index()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_index.html)

````python
# When we compare to the histogram above, we can see whether our stratified sampling worked or not.

strat_test_set['income_cat'].value_counts().sort_index() / len(strat_test_set)
````

### After we've conducted stratified sampling, we need to remove the `income_cat` column  because it's of no use to us anymore.

````python
for set_ in (strat_train_set, strat_test_set):
    set_.drop('income_cat', axis=1, inplace=True)
    
print(strat_train_set.describe())
````

## 4-2. Looking for Correlations

### Since our dataset is not very large, we can just compute the [Pearson correlation coefficient](https://en.wikipedia.org/wiki/Pearson_correlation_coefficient) between every pair of attributes using the [`pandas.DataFrame.corr()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html) method.

### Let's first calculate the correlation for each attribute paired up with `median_house_value` and also draw a heatmap matrix to see the correlation of each variable paired with each other variable using [`matplotlib.axes.Axes.matshow()`](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.matshow.html)

#### Reference:

1. Pearson correlation coefficient:

  * In statistics the Pearson correlation coefficient (PCC or Pearson's r) measures the correlation between two variables. It takes the covariance between the two variables ($\text{Cov}(X,\ Y)$) and divides that by the product of the standard deviations of both variables ($\sqrt{\text{Var}(X)\text{Var}(Y)}$ or $\sigma(X)\sigma(Y)$).
  * The covariance is computed as follows: $\text{Cov}(X,\ Y) = \text{E}(XY) - \text{E}(X)\text{E}(Y)$
  
  $$\rho(X,\ Y) = \frac{\text{Cov}(X,\ Y)}{\sigma(X)\sigma(Y)}$$
  
2. [`pandas.DataFrame.corr()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html)

3. [`matplotlib.axes.Axes.matshow()`](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.matshow.html)

````python
corr_matrix = housing_copy.corr()
print(corr_matrix.head())
````

````python
# Columns of DataFrame except for categorical value columns.
def housing_heatmap(data):
    columns = data.select_dtypes(exclude='object').columns
    corr_matrix = data.corr()

    fig, ax = plt.subplots(figsize=(8, 8))
    mat = ax.matshow(corr_matrix, cmap='jet')

    ax.set_xticks(range(len(columns)))
    ax.set_yticks(range(len(columns)))
    ax.set_xticklabels(columns)
    ax.set_yticklabels(columns)
    plt.setp(ax.get_xticklabels(), rotation=45, ha='left', rotation_mode='anchor')
    plt.colorbar(mat, fraction=0.045, pad=0.05)

    fig.tight_layout()
    plt.show()
    return None
    
    
def house_value_corr(data=housing_copy):
    corr_matrix = data.corr()
    print("Corr(median_house_value, X)")
    print('-' * 30)
    print(corr_matrix['median_house_value'].sort_values(ascending=False))
    return None
    
    
housing_heatmap(housing_copy)
print('\n')
house_value_corr()
````

# 4-3. Experimenting with Attribute Combinations

### We've taken a look at a lot of stuff, but we may also want to consider combining attributes with one another. For example, the `total_rooms` attribute is not very useful in and of itself. What we want to know is how many rooms there are per household. Also, we want to know how many bedrooms there are compared to the total number of rooms, not just how many bedrooms there are.

#### This process can be further called _**feature engineering**_. There are entire books written on this topic and some people go as far as to saying that machine learning is just a matter of how much data you have and how many relevant features you can create.

### Let's create our new attributes and see the correlation again.

<br>

````python
housing_copy['rooms_per_household'] = housing_copy['total_rooms'] / housing_copy['households']
housing_copy['bedrooms_per_rooms'] = housing_copy['total_bedrooms'] / housing_copy['total_rooms']
housing_copy['population_per_household'] = housing_copy['population'] / housing_copy['households']
````

````python
housing_heatmap(housing_copy)
print('\n')
house_value_corr()
````

## 5-1. Data Cleaning

### This section focuses on cleaning up the missing features that we saw earlier. Most machine learning algorithms cannot work with missing features, so it's important that we take care of these.

### There are a few ways that we can do this.

#### 1. Get rid of the samples that have missing values (listwise deletion a.k.a. complete-case analysis).

#### 2. Get rid of the whole attribute.

#### 3. Set the missing values to some value (mean, median, regression, etc.).

### Pandas actually has built-in functions that let us do these operations conveniently: [`pandas.DataFrame.dropna()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html), [`pandas.DataFrame.drop()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html), and [`pandas.DataFrame.fillna()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html).
